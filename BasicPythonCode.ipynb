```python
def is_prime(n):
    """
    Ex.1: Checks if a number is a prime number.
    A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
    """
    # Prime numbers must be greater than 1.
    if n <= 1:
        return False
    # Check for divisors from 2 up to n-1.
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


print(f"Is 7 a prime number? {is_prime(7)}") # Expected output: True
print(f"Is 10 a prime number? {is_prime(10)}") # Expected output: False
```

```python
import random

def product_of_randoms():
    """
    Ex.2 : Generates two random numbers, asks the user for their product, 
    and checks if the answer is correct.
    """
    #Its going to generate two random integers between 1 and 100.
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    
    #Calculate the correct product.
    correct_product = num1 * num2
    
    #Ask the user for their answer.
    try:
        user_answer = int(input(f"What is the product of {num1} and {num2}? "))
        
        # Check if the users answer is correct.
        if user_answer == correct_product:
            print("Correct!")
        else:
            print(f"Incorrect. The correct answer is {correct_product}. Better luck next time!")
            
    except ValueError:
        print("Invalid input. Please enter a number.")

# Run the function.
product_of_randoms()
```

```python
# Ex.3: Squares of Even/Odd Numbers
# Prints the squares of all even and odd numbers in the range 100 to 200.

print("Squares of Even Numbers (100-200):")  # Loop through numbers from 100 to 200
for num in range(100, 201):   
    if num % 2 == 0:     # Check if the number is even using the modulo operator      
        print(f"{num}^2 = {num**2}")    # Print the number and its square

print("\nSquares of Odd Numbers (100-200):")
# Loop through numbers from 100 to 200
for num in range(100, 201):
    # Check if the number is odd
    if num % 2 != 0:
        print(f"{num}^2 = {num**2}")   # Print the number and its square
```

```python
# Ex.4: Word Counter
# Counts the frequency of each word in a given text.

# We'll use the 'collections' module to make counting easier.
from collections import Counter
import re

def word_counter(text):
    # Clean the text: convert to lowercase and find all words.
    words = re.findall(r'\b\w+\b', text.lower())
    
    # Use Counter to get the frequency of each word.
    word_counts = Counter(words)
    
    return word_counts

# Ex text from the assignment.
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Get the word counts and print the result.
counts = word_counter(input_text)
print(counts)
```

```python
# Ex.5: Check for Palindrome
# A palindrome is a word or phrase that reads the same forwards and backwards.

def is_palindrome(text):
    # Clean the input text: convert to lowercase and remove non-alphanumeric characters.
    # The `isalnum()` method checks if a character is alphanumeric (A-Z, a-z, and 0-9).
    cleaned_text = ''.join(char for char in text if char.isalnum()).lower()
    
    # Compare the cleaned text to its reverse.
    # `[::-1]` is a Python slice that reverses a string.
    return cleaned_text == cleaned_text[::-1]

# Example from the assignment.
print(f"Is 'racecar' a palindrome? {is_palindrome('racecar')}")  # Expected: True
print(f"Is 'A man, a plan, a canal: Panama.' a palindrome? {is_palindrome('A man, a plan, a canal: Panama.')}")  # Expected: True
print(f"Is 'hello' a palindrome? {is_palindrome('hello')}")      # Expected: False
```
